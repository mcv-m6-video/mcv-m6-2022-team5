```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from optical_flow import *
```

```python
data_path = '../datasets/results_opticalflow_kitti'
```

## Image 45

```python
gt_flow = load_flow(os.path.join(data_path,'gt_000045_10_noc.png'))
kitti_flow = load_flow(os.path.join(data_path,'LKflow_000045_10.png'))
```

```python
plot_optical_flow(gt_flow)
plot_optical_flow(kitti_flow)
```

```python
mse = flow_msen(gt_flow, kitti_flow)
print('Mean Square Error in Non-occluded areas: ', mse)
```

```python
pepn = flow_pepn(gt_flow, kitti_flow)
print('Percentage of Erroneous Pixels in Non-occluded areas: ',str(pepn) +'%')
```

```python
error_dis = flow_error_distance(gt_flow,kitti_flow)
plot_error_distance(error_dis)
```

```python
plot_error_distribution_hist(error_dis,gt_flow)
```
